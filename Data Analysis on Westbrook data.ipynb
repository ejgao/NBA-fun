```python
#data analysis
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\Users\\Eric Gao\\Downloads\\Webscraping\\players.csv",header=1)
pd.set_option('display.max_columns', 40)
df.columns = df.columns.str.strip()
#sns.lmplot("FTM", "FTA", data=df)
#plt.scatter(df.FTM, df.FTA, color="orange")
#Linear relationship between free throw attempts and free throws made
#Finding a correlation between all the factors
extract=df.corr(method="pearson")
#only want to print with strong correlations and only want to test with these values(r>=0.5 or <=-0.5)
pos=extract[extract>=0.5]
neg=extract[extract<=-0.5]
pos
#R is >0.5 for effective field goal%(EFG%), 3P%, 2P%, and FT%. Use multivariate regression.
newdf=pd.DataFrame(pos)
newdf=newdf.fillna(0)
#0's are areas where pos correlation is below 0.5.
newdf
```

```python
#Factors that impact EFG%--FGM, FGA, 3P, 3PA, 2P, 2P%, Steals, Points, and Oreb(3P+2P=FGM, 3PA+2PA=FGA, however, 2PA does not seem to 
#have a strong positive impact)
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import scale
import numpy as np
model=LinearRegression()
x=df[["Age","FGM", "FGA", "3P", "3PA", "2P", "2P%", "Steal", "Points", "OReb"]]
y=df["EFG%"]
X_train, X_test, y_train, y_test=train_test_split(x,y,test_size=0.2)
model.fit(X_train, y_train)
#Accuracy of model
#Negative score- maybe scaling needed?
np.mean(cross_val_score(LinearRegression(), X_train, y_train))
```
