```python
# +/-: Plus minus is used to indicate a player's impact on a game, i.e., when the player was in the game, how many points
# did the opponent score compared to that player's team. If Lebron scores 2 points, thats +2. However, if he opponent 
# scores 3 points, that would be -1. For more details, visit the ReadMe.
df1=df[df.Series==FIN]
df2=df1[df1.Opponent != "DAL"]
df3=df1[df1.Opponent == "DAL"]
final_plus_minus=df1["+/-"].mean()
plus_minus_no_dal=df2["+/-"].mean()
dal_plus_minus=df3["+/-"].mean()
plus_minus_std=df1["+/-"].std()
print("Lebron's average plus minus across all NBA Finals games is: " + str(round(final_plus_minus, 2)) + " .")
print("Lebron's average plus minus across all NBA Finals without the Dallas series is: " + str(round(plus_minus_no_dal,2)) + " .")
print("Lebron's average plus minus in the Dallas Series is: " + str(round(dal_plus_minus, 2)) + " .")
print("Standard deviation of Lebron's plus minus in all NBA Finals is: " + str(round(plus_minus_std, 2)) + " .")
print("Lebron's plus minus during the Dallas series was still within one standard deviation of his average across all Finals.")
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
fig = make_subplots(rows=1, cols=2)

fig.add_trace(
    go.Scatter(x=df2.Date, y=df2["+/-"]),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(x=df3.Date, y=df3["+/-"]),
    row=1, col=2
)

fig.update_layout(
    title_text=("+/- excluding the Dallas Series(left) and during(right)")
)
fig.update_xaxes(title_text="Date", row=1, col=1)
fig.update_xaxes(title_text="Date", row=1, col=2)
fig.update_yaxes(title_text="+/-")
fig.update_yaxes(range=[-50,50])

fig.show()
#Lebron's plus minuses across all finals is higher than his plus minuses during the Dallas series, as indicated by the chart.
#Values on thhe chart on the right are on the lower end.
```
