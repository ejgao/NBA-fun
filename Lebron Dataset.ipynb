```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\Users\\Eric Gao\\Downloads\\Webscraping\\lebron.csv",skiprows=1)
df.columns=df.columns.str.strip()
df=df.fillna(0)
pd.set_option('display.max_columns', 400)
f = plt.figure(figsize=(10,7))
plt.plot(df.Date, df.Points)
plt.xlabel("Date")
plt.ylabel("Points")
plt.title("Points per playoff game")
#Career Playoff Points in every game
# Seems like most of points in between 20-35 range.
df.Points.mean()
# Lot of highs and lows
# Career high in a finals game on 5/31/18 wih 51
# Career low on 6/7/2011 with 8
```

```python
f = plt.figure(figsize=(10,7))
a=df[df.Series=="FIN"]
plt.plot(a.Date, a.Points)
plt.title("Points per Finals Game")
plt.xlabel("Date")
plt.ylabel("Points")
plt.show() #Points per game in each NBA finals game that Lebron has played in.
```

```python
#Career finals average
points=df[df.Series=="FIN"]
newdf=points[df.Opponent != "DAL"]
no_dallas=newdf.Points.mean() # Without Dallas Series, points average in the NBA finals is 29.67.
mean_points=df[df.Series=="FIN"].Points.mean()
std_points=df[df.Series=="FIN"].Points.std()
print("NBA finals average is: " + str(round(mean_points, 2)) + " points with a standard deviation of "+ str(round(std_points,2)) + " points.")
print("Excluding the Dallas series, his career average increases to " + str(round(no_dallas, 2)) + " points. ")
```

```python
#Lebron's career in the NBA finals is 28.22 points. However, his performance in the 2011 NBA Finals against the Dallas Mavericks stands out infamously,
#as shown in the above cell and the above graph.
#On 6/7/2011, he scored 8 points, which is the lowest he has ever scored in an NBA finals game.
#Find average in the 2011 finals.
a=df[df.Series=="FIN"]
dallas_average=a[a.Opponent=="DAL"].Points.mean()
print("Lebron's average in the 2011 NBA Finals against Dallas was:",str(round(dallas_average, 2)) + " points.")
#Lebron's average points per game in the Dallas series is 12 points lower than his average points per game wihout taking
#into account the Dallas series. Furthermore, his average points per game in the Dallas series is more than one standard deviation
#away from his career average.
#What factors could have caused this? I will be using the data scraped and provided in the CSV to analyze.
```

```python
df1=df[df.Series=="FIN"]
average_fga=df.FGA.mean()
df2=df1[df1.Opponent != "DAL"]
average_no_dal=df2.FGA.mean()
df3=df1[df1.Opponent == "DAL"]
```

```python
f = plt.figure(figsize=(10,7))
plt.subplot(1,2,1)
plt.plot(df2.Date, df2.Points)
plt.xlabel("Date")
plt.ylabel("Points")
plt.title("Points in the finals excluding Dallas series")
plt.subplot(1,2,2)
plt.plot(df3.Date, df3.Points)
plt.xlabel("Date")
plt.ylabel("Points")
plt.title("Points in the finals during Dallas series")
#Lebron's average points was significantly lower in the Dallas series than any other finals series, and him
#getting less points is a possible factor into his poor performance in the 2011 Finals.
```

```python
#Field Goal Average
df1=df[df.Series=="FIN"]
average_fga=df.FGA.mean()
df2=df1[df1.Opponent != "DAL"]
average_no_dal=df2.FGA.mean()
df3=df1[df1.Opponent == "DAL"]
average_dal=df3.FGA.mean()
print("Lebron's average field goal attempts per game in his career in the NBA finals is: "+str(round(average_fga,2)) + " attempts.")
print("Excluding the Dallas series, his average field goals attempted is: "+ str(round(average_no_dal,2)) + " attempts and his average field goals attempted in Dallas series is: "+ str(round(average_dal,2)) + " attempts.")
```

```python
f = plt.figure(figsize=(12,7))
plt.subplot(1,2,1)
plt.plot(df2.Date, df2.FGA)
plt.xlabel("Date")
plt.ylabel("Field Goal Attempts")
plt.title("Field Goal Attempts in the finals excluding Dallas series")
plt.subplot(1,2,2)
plt.plot(df3.Date, df3.FGA)
plt.xlabel("Date")
plt.ylabel("Field Goal Attempts")
plt.title("Field Goal Attempts in the finals during Dallas series")
#Lebron's average field goals attempted was significantly lower in the Dallas series than any other finals series.
```

```python
#Field goal percentage
df1=df[df.Series=="FIN"]
df2=df1[df1.Opponent != "DAL"]
df3=df1[df1.Opponent == "DAL"]
fg_percent_finals=df1["FG%"].mean()
fg_percent_finals_no_dal=df2["FG%"].mean()
fg_percent_dal=df3["FG%"].mean()
print("Career playoff field goal percentage: " + str(round(fg_percent_finals*100,2)) + "%.")
print("Career playoff field goal percentage: " + str(round(fg_percent_finals_no_dal*100,2)) + "%.")
print("Playoff field goal percentage in Dallas series: "+ str(round(fg_percent_dal*100,2)) + "%.")
#print("Standard deviation of career playoff field goal percentage is:",df1[["FG%"]].std()) #dallas series lies within one standard deviation. 
```

```python
f = plt.figure(figsize=(10,7))
plt.subplot(1,2,1)
plt.plot(df2.Date, df2["FG%"])
plt.xlabel("Date")
plt.ylabel("Field Goal %")
plt.title("Field Goal % in the finals excluding Dallas series")
plt.subplot(1,2,2)
plt.plot(df3.Date, df3["FG%"])
plt.xlabel("Date")
plt.ylabel("Field Goal %")
plt.title("Field Goal % in the finals during Dallas series")
#Field goal percentages across all NBA Finals games by Lebron James are actually very consistent with one another.
```

```python
#Compare GmSc averages. GmSc is a measure of productivity. Link is provided in the ReadMe.
df1=df[df.Series=="FIN"]
df2=df1[df1.Opponent != "DAL"]
df3=df1[df1.Opponent == "DAL"]
game_score_finals=df1["GmSc"].mean()
game_score_nodal=df2["GmSc"].mean()
game_score_dal=df3["GmSc"].mean()
print("Lebron's average game score in the NBA Finals is: ",round(game_score_finals,2))
print("Lebron's average game score in the NBA Finals excluding the Dallas series was:",round(game_score_nodal,2))
print("Lebron's average game score in the Dallas series was: ", round(game_score_dal,2))
print("Lebron's game score in Dallas was nearly 10 points lower than his career average game score in the finals.")
```

```python
f = plt.figure(figsize=(10,7))
plt.subplot(1,2,1)
plt.plot(df2.Date, df2["GmSc"])
plt.xlabel("Date")
plt.ylabel("Game score")
plt.title("Game Score in the finals excluding Dallas series")
plt.subplot(1,2,2)
plt.plot(df3.Date, df3["GmSc"])
plt.xlabel("Date")
plt.ylabel("GmSc")
plt.title("Game score in the finals during Dallas series")
#Lebron's game score in all finals excluding the Dallas series is significantly higher than his score in the Dallas series.
```

```python
# +/-: Plus minus is used to indicate a player's impact on a game, i.e., when the player was in the game, how many points
# did the opponent score compared to that player's team. If Lebron scores 2 points, thats +2. However, if he opponent 
# scores 3 points, that would be -1. For more details, visit the ReadMe.
df1=df[df.Series=="FIN"]
df2=df1[df1.Opponent != "DAL"]
df3=df1[df1.Opponent == "DAL"]
final_plus_minus=df1["+/-"].mean()
plus_minus_no_dal=df2["+/-"].mean()
dal_plus_minus=df3["+/-"].mean()
plus_minus_std=df1["+/-"].std()
print("Lebron's average plus minus across all NBA Finals games is: " + str(round(final_plus_minus, 2)) + " .")
print("Lebron's average plus minus across all NBA Finals without the Dallas series is: " + str(round(plus_minus_no_dal,2)) + " .")
print("Lebron's average plus minus in the Dallas Series is: " + str(round(dal_plus_minus, 2)) + " .")
print("Standard deviation of Lebron's plus minus in all NBA Finals is: " + str(round(plus_minus_std, 2)) + " .")
print("Lebron's plus minus during the Dallas series was still within one standard deviation of his average across all Finals.")
```

```python
f = plt.figure(figsize=(10,7))
plt.subplot(1,2,1)
plt.plot(df2.Date, df2["+/-"])
plt.xlabel("Date")
plt.ylabel("+/-")
plt.title("+/- in the finals excluding Dallas series")
plt.subplot(1,2,2)
plt.plot(df3.Date, df3["+/-"])
plt.xlabel("Date")
plt.ylabel("+/-")
plt.title("+/- in the finals during Dallas series")
#Lebron's plus minuses across all finals is higher than his plus minuses during the Dallas series, as indicated by the chart.
#Values on thhe chart on the right are on the lower end.
```

```python
#Next few cells will test the impact of the amount of minutes played
time=[]
for element in df["Min played"]:
    element=element[:5]
    element=element.replace(":",".")
    time.append(float(element))
df["Min played"]=time
```

```python
df1=df[df.Series=="FIN"]
df2=df1[df1.Opponent != "DAL"]
df3=df1[df1.Opponent == "DAL"]
min_played_finals=df1["Min played"].mean()
min_played_finals_nodal=df2["Min played"].mean()
dal_min=df3["Min played"].mean()
print("In Lebron's NBA Final's career, he averaged: "+ str(round(min_played_finals,2)) + " minutes.")
print("In every other NBA Finals Series(excluding Dallas), Lebron has averaged: " + str(round(min_played_finals_nodal,2)) + " minutes.")
print("In the Dallas series, Lebron averaged: " + str(round(dal_min,2)) + " minutes.")
```

```python
f = plt.figure(figsize=(10,7))
plt.subplot(1,2,1)
plt.plot(df2.Date, df2["Min played"])
plt.xlabel("Date")
plt.ylabel("Min played")
plt.title("Min played in the finals excluding Dallas series")
plt.subplot(1,2,2)
plt.plot(df3.Date, df3["Min played"])
plt.xlabel("Date")
plt.ylabel("Min played")
plt.title("Min played in the finals during Dallas series")
#In terms of minutes played, Lebron actually averaged around a minute more in the Dallas series, so the element of 
#fatigue is an unlikely factor. Chart verifies how his minutes played across all NBA finals are very similar.
```
